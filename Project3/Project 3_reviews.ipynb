```python
stop = stopwords.words('english')
```

```python
stemmer = PorterStemmer()

def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

def tokenize(text):
    translator = str.maketrans('', '', string.punctuation)
    no_punctuation = text.translate(translator)
    tokens = nltk.word_tokenize(no_punctuation)
    stems = stem_tokens(tokens, stemmer)
    return stems
```

```python
len(text_data)
```

```python
tfidfvector = TfidfVectorizer(tokenizer=tokenize, ngram_range = (1,1), min_df = 0.15, max_df = 0.95, stop_words=stop)
# tfidfvector = TfidfVectorizer(ngram_range = (1,1),min_df = 0.3, max_df = 0.95, stop_words=stop)
text_tfidf = tfidfvector.fit_transform(text_data)
text_tfidf.shape
```

```python
wordbag = tfidfvector.get_feature_names()
```

```python
wordbag
```

```python
text_df = text_tfidf.toarray()
```

```python
review = pd.DataFrame(text_df, columns=wordbag)
```

```python
review
```

```python
modi_review = review.copy()
```

```python
for word in wordbag:
    modi_review.ix[modi_review[word] == 0, word] = 0
    modi_review.ix[modi_review[word] != 0, word] = 1
```

```python
for s in wordbag:
    modi_review[s] = modi_review[s].astype(int)
```
