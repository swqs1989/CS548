```python
df_data = pd.read_table("processed.cleveland.data", 
                        sep=',', 
                        names=["age","sex","cp","trestbps","chol","fbs","restecg",
                               "thalach","exang","oldpeak","slope","ca","thal","num"])
```

```python
# check NaN in the data
for feature in df_data.columns:
    if sum(pd.isnull(df_data[feature])) != 0 or sum(df_data[feature].isin(["?"])):
        print("%r column: %r NaN" %(feature, sum(pd.isnull(df_data[feature]))))
        print("%r column: %r ?" %(feature, sum(df_data[feature].isin(["?"]))))
```

```python
# Here, I use the mode of that attribute to replace the question markers.
modeThal = df_data["thal"].mode()
modeThal = modeThal[0]
```

```python
modeCa = df_data["ca"].mode()
modeCa = modeCa[0]
```

```python
df_data["thal"] = df_data["thal"].replace("?", modeThal)
```

```python
df_data["ca"] = df_data["ca"].replace("?", modeCa)
```
