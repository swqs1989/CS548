```python
def clean_data(df):
    df = df.replace('N/A', np.NaN)
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].str.replace("$", "")
            df[col] = df[col].str.replace(",", "")
            try:
                df[col] = pd.to_numeric(df[col])
            except:
                df[col] = df[col].astype(str) 
        if sum(pd.isnull(df[col])) != 0 or sum(df[col].isin(["?"])) > 0:
            if df[col].dtype == "object":
                md = df[df[col] != np.NaN][col].mode()[0]
                df[col] = df[col].replace(np.NaN, md)
                
            else:
                mn = df[col].astype(float).median()
                df[col] = df[col].replace(np.NaN, mn)
        if df[col].dtype == "object":
            df[col] = df[col].astype(str)
    return df
```

```python
def standardize(narray):
    mean = np.mean(narray, axis = 0)
    std = np.std(narray, axis = 0)
    stdized = (narray - mean)/std
    if np.isnan(np.sum(stdized)):
        print("NAN showed!")
        stdized = np.nan_to_num(stdized)
    return stdized
```

```python
def visualManifold(arr, clusters):
    methods = ['standard', 'ltsa', 'hessian', 'modified']
    labels = ['LLE', 'LTSA', 'Hessian LLE', 'Modified LLE']
    n_neighbors = 10
    fig = plt.figure(figsize=(20, 10))
    for i, method in enumerate(methods):
        tr = manifold.LocallyLinearEmbedding(n_neighbors, 2,
                                    method=method).fit_transform(arr)
        ax = fig.add_subplot(252 + i)
        plt.scatter(tr[:,0], tr[:, 1], cmap=plt.cm.rainbow, c = clusters)
        plt.title("%s" % (labels[i]))
        plt.axis('tight')
    mds = manifold.MDS(n_components = 2, max_iter= 500, metric = True, n_init= 10).fit_transform(arr)
    ax = fig.add_subplot(257)
    plt.scatter(mds[:,0], mds[:, 1], c = clusters, cmap=plt.cm.rainbow)
    plt.title("MDS")

    iso =  manifold.Isomap(n_components = 2, n_neighbors= n_neighbors).fit_transform(arr)
    ax = fig.add_subplot(258)
    plt.scatter(iso[:,0], iso[:, 1], c = clusters, cmap=plt.cm.rainbow)
    plt.title("Isomap")

    iso =  manifold.SpectralEmbedding(n_components = 2, n_neighbors= n_neighbors).fit_transform(arr)
    ax = fig.add_subplot(259)
    plt.scatter(iso[:,0], iso[:, 1], c = clusters, cmap=plt.cm.rainbow)
    plt.title("SpectralEmbedding")

    tsne = manifold.TSNE(n_components=2, init='pca', random_state=0).fit_transform(arr)
    ax = fig.add_subplot(2, 5, 10)
    plt.scatter(iso[:,0], iso[:, 1], c = clusters, cmap=plt.cm.rainbow)
    plt.title("TSNE")
```

```python
def add_labels(arr, label):
    new_arr = np.hstack((arr, label))
    return new_arr
```

```python
def visualPCA(data, targets):
    pca = PCA(n_components = 3)
    pca.fit(data)
    X = pca.transform(data)
    fig = plt.figure(1, figsize=(4, 3))
    plt.clf()
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    plt.cla()
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=targets, cmap=plt.cm.rainbow)
```

## Q1 Scipy Hierachy Cluster

```python
majors = df_degree.iloc[:, 0].values
degree_nume =df_degree.iloc[:, 1:]
degree_stdized = standardize(degree_nume.values)
df_stdized = pd.DataFrame(degree_stdized, columns = degree_nume.columns)
```

```python
%matplotlib inline  
import scipy.cluster.hierarchy as sch
fig = pylab.figure(figsize=(8,8))
Y = sch.linkage(degree_stdized, method='single', metric = 'euclidean', )
# format: [idx1, idx2, dist, sample_count].
Z = sch.dendrogram(Y, orientation='right', labels = majors)
clusters = sch.fcluster(Y, 1.5, criterion = "distance", depth = 3)
clusters
```

```python
%matplotlib

visualPCA(degree_stdized, clusters)
```

```python
%matplotlib inline  

visualManifold(degree_stdized, clusters)
```

```python
%matplotlib inline  
    
visualParallel(df_stdized, clusters)
```
