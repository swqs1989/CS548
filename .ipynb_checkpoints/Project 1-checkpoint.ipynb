```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

predictors = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year"]

X = df_auto_str[predictors]
y = df_auto_str["origin"]

logistic = LogisticRegression()
rfe = RFE(logistic, 2)
log_fit = rfe.fit(X, y)
```

```python
log_fit.n_features_
```

```python
log_fit.support_
```

```python
log_fit.ranking_
```

```python
for i in range(1, 5):
    logistic = LogisticRegression()
    rfe = RFE(logistic, i)
    log_fit = rfe.fit(X, y)
    print("n_features_: %r" %log_fit.n_features_)
    print("support_: %r" %log_fit.support_)
    print("ranking_: %r" %log_fit.ranking_)
    print("Selected feature name: %r" %X.columns[log_fit.support_])
    print("="*100)
```
